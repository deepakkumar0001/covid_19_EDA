```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

```

```python
new_df=pd.read_csv("/content/covid-variants.csv")
```

#Cleaning of data

```python
new_df.replace(to_replace=0, value=np.NaN, inplace=True)
new_df
```

```python
df=new_df.dropna()
df.head(20)
```

#Reset of index

```python
df.reset_index(level=None, drop=False, inplace=True,)
df.style.background_gradient(cmap='Reds')
```

```python
df.describe()
```

```python
df['variant'].unique()
```

#Q.1= Which type of variant register highested time in a particular month ?

```python
df['date_new']=pd.to_datetime(df['date'])
```

```python
df.head(2)
```

```python
df_variant=df['date_new'].dt.month.value_counts().to_frame()
df_variant.style.background_gradient(cmap='Greens')
```

here we find that 5th month have highest variant registered 

```python
df['date_new'].dt.month.value_counts().plot(kind='bar')
plt.rcParams["figure.figsize"] = (15,8)
plt.title("Month wise total variant")
```

#Q.2= Which variant register most ?

```python
df_variant_type=df['variant'].value_counts().to_frame()
df_variant_type.style.background_gradient(cmap='Wistia_r')
```

```python
df['variant'].value_counts().plot(kind='bar')
plt.rcParams["figure.figsize"] = (10,9)
plt.title("Highest registered variant")
```

Here we can see that which variant is highest time registered

#Q.3= find out which country have highest num_sequence with variant type ?

```python
last_df=df.groupby(["location","variant"])['num_sequences'].sum()
```

```python
last_dfs=last_df.sort_values(ascending=False).head(30).to_frame()
last_dfs.style.background_gradient(cmap='Reds')
```

Here we can get that which country have highest time of num_sequences in which variant

#Q.4=Find highest time of num_sequence register on particular date and location ,  which type of variant is that register? 

```python
data_df = df.groupby(['date_new'])['location','num_sequences_total','variant'].max()
```

```python
data_df1=data_df.sort_values(by=["num_sequences_total",'variant'],ascending=False).head(20)
data_df1.style.background_gradient(cmap="Wistia_r")
```

Here we can see that a which type of variant register with highest num_sequence  on particular date and in which country
